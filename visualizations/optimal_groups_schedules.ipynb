## Import Packages

```python
import sys
sys.path.insert(0, "../utils")

import schedules
import utils_functions
from unavailability import optimal_group1_chairs, optimal_group1_members, optimal_group1_unavailability, optimal_group2_chairs, optimal_group2_members, optimal_group2_unavailability
```

## Import Optimal Data

```python
prefix_path = "../data/optimal_groups"
group1_optimal = utils_functions.read_optimal_txt(f"{prefix_path}/group1_results.txt")
group2_optimal = utils_functions.read_optimal_txt(f"{prefix_path}/group2_results.txt")
```

```python
rooms_per_timeslot = schedules.map_timeslots_to_num_rooms(group1_optimal, group2_optimal)
```

## Save Schedules

```python
group1_people = [*optimal_group1_chairs, *optimal_group1_members]
group2_people = [*optimal_group2_chairs, *optimal_group2_members]

schedules.save_group_schedules(group1_people, group1_optimal, optimal_group1_unavailability, prefix_path)
schedules.save_group_schedules(group2_people, group2_optimal, optimal_group2_unavailability, prefix_path)
```
